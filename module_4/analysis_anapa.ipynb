## Пояснение к датасету

  1. flight_id - уникальный номер рейса
  2. flight_no - неуникальный номер рейса
  3. aircraft_code - код самолёта, совершившего рейс
  4. departure_airport - код аэропорта вылета в Анапе
  5. arrival_airport - код аэропорта прилёта
  6. scheduled_departure - время вылета из Анапы по расписанию
  7. actual_departure - реальное время вылета рейса из Анапы
  8. scheduled_arrival - время прилёта в аэропорт назначения по расписанию
  9. actual_arrival - реальное время прилёта в аэропорт назначения
  10. flight_time - время полёта в часах
  11. economy_count - количество проданных билетов эконом-класса на рейс
  12. economy_income - суммарная стоимость проданных билетов эконом-класса на рейс в рублях
  13. business_count - количество проданных билетов бизнес-класса на рейс
  14. business_income - суммарная стоимость проданных билетов бизнес-класса на рейс в рублях
  15. total_count - суммарное количество проданных билетов на рейс
  16. total_income - суммарная стоимость проданных билетов на рейс в рублях
  17. fuel_consumption - потраченное топливо на рейс в тоннах
  18. fuel_cost - затраты на топливо в рублях
  19. revenue - выручка от рейса в рублях


```python
data = pd.read_csv('anapa_air_2017.csv')
data
```

```python
data.info()
```

```python
# Количество направлений
data['arrival_airport'].value_counts()
```

```python
# Отсутствует информация и пассажирах в направлении NOZ(Новокузнецк) 13 рейсов.
data[data.arrival_airport == 'NOZ']
```

```python
# Делаем второй датасет, в котором убираем данные о рейсах в Новокузнецк, поскольку о них нет данных
df = data[data.arrival_airport != 'NOZ'].copy()
```

```python
# Обрабатываем столбец со временем отправления. Достаём даты из текстовых строк
df['actual_departure'] = df['actual_departure'].apply(lambda s: s[:10])
df['actual_departure'] = df['actual_departure'].apply(
    lambda t: datetime.strptime(t, '%Y-%m-%d'))
```

```python
# Функция, меняющая номер дня на его название
def weekday_name(n):
    if n == 0:
        return 'Monday'
    if n == 1:
        return 'Tuesday'
    if n == 2:
        return 'Wednesday'
    if n == 3:
        return 'Thursday'
    if n == 4:
        return 'Friday'
    if n == 5:
        return 'Saturday'
    if n == 6:
        return 'Sunday'


# Добавляем столбцы с номером месяца, номером дня и днём недели вылета
df['month'] = df['actual_departure'].apply(lambda x: x.month)
df['day'] = df['actual_departure'].apply(lambda y: y.day)
df['weekday'] = df['actual_departure'].apply(lambda z: z.weekday())
df['weekday'] = df['weekday'].apply(weekday_name)
df['week'] = df['actual_departure'].apply(lambda z: z.isocalendar()[1])
df['week'][df['week'] == 52] = 1
```

```python
# Разделяем датасет на два для разных самолётов
df_SU = df[df['aircraft_code'] == 'SU9'].copy()
df_733 = df[df['aircraft_code'] == '733'].copy()
```

## Первый способ отбора рейсов.

 Самый простой. Отменяем все рейсы ниже какого-то значения прибыльности. Например, рейсы, выручка которых меньше 0.95 от средней выручки по всем рейсам в этот город.

```python
# Рейсы Boeing 737-300
df_733[df_733['revenue'] < 0.95 * df_733['revenue'].mean()][['flight_id',
                                                             'aircraft_code', 'actual_departure', 'revenue']].sort_values(by='revenue')
```

```python
# Рейсы Sukhoi Superjet-100
df_SU[df_SU['revenue'] < 0.95 * df_SU['revenue'].mean()][['flight_id',
                                                          'aircraft_code', 'actual_departure', 'revenue']].sort_values(by='revenue')
```

Значение об прибыльности можно установить только при получении всей информации о затратах и прибыли от полёта.

## Третий способ отбора рейсов
Такой же, как и второй, только сортировка идёт по номеру недели в году. Будем отменять самые неприбыльные рейсы в течение каждой недели. Допустим, отменим все рейсы, доход от которых ниже первого квартиля.

Рейсы Boeing 737-300

```python
sns.boxplot(x='week', y='revenue', data=df_733)
```

```python
# Вспомогательная таблица, считающая 1-й квантиль для каждой недели
W_733 = df_733.groupby(['week'])['revenue'].quantile(0.25)
W_733
```

```python
# Рейсы, прибыль от которых меньше 1-го квартиля
for i in W_733.index:
    display(df_733[df_733['week'] == i][df_733['revenue'] < W_733[i]][[
            'flight_id', 'aircraft_code', 'actual_departure', 'revenue']])
```

Рейсы Sukhoi Superjet-100

```python
sns.boxplot(x='week', y='revenue', data=df_SU)
```

```python
# Вспомогательная таблица, считающая 1-й квантиль для каждой недели
W_SU = df_SU.groupby(['week'])['revenue'].quantile(0.25)
W_SU
```

```python
# Рейсы, прибыль от которых меньше 1-го квартиля
for j in W_SU.index:
    display(df_SU[df_SU['week'] == j][df_SU['revenue'] < W_SU[j]]
            [['flight_id', 'aircraft_code', 'actual_departure', 'revenue']])
```

## Итоговая рекомендация
Так как для каждого способа получилось довольно много рейсов для отмены, примем решение отменить только те рейсы, которые были найдены во всех трёх способах.

```python
# Для Boeing 737-300 это следующие рейсы:
Cancel_733 = [
    136122,
    136130,
    136178,
    136185,
    136253,
    136282,
    136306,
    136352,
    136360,
    136397,
    136420,
    136464
]

# Для Sukhoi Superjet-100 это следующие рейсы:
Cancel_SU = [
    136620,
    136642,
    136678,
    136800,
    136807,
    136823,
    136843,
    136844,
    136887,
    136888,
    136922,
    136929,
    136937
]
```

```python
# Создаём таблицу с отменёнными рейсами:
df_cancel = df.copy()
df_cancel

# И убираем из них выбранные рейсы
for n in Cancel_733:
    df_cancel.drop(df_cancel[df_cancel['flight_id']
                             == n].index, axis=0, inplace=True)

for m in Cancel_SU:
    df_cancel.drop(df_cancel[df_cancel['flight_id']
                             == m].index, axis=0, inplace=True)
```

```python
# Считаем среднюю и медианную прибыльность с рейса начальной таблицы
df.groupby(['aircraft_code'])['revenue'].agg(['mean', 'median'])
```

```python
# Считаем среднюю и медианную прибыльность с рейса итоговой таблицы
df_cancel.groupby(['aircraft_code'])['revenue'].agg(['mean', 'median'])
```
